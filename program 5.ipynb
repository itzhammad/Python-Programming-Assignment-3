```python
#5. Write a Python program which accepts the user's first and last name and print them in reverse order with a space between them.

def reverse(name):
    rev_name = ""
    for i in name:
        rev_name = i + rev_name               #python reads the code line by line from left to right
    return rev_name                           
    
fname, sname = input(),input()
name = fname+" "+sname
print("Original Name: "+name+"\nName in reverse order: "+reverse(name))
```
