```python
#11.Write a Python program to get the largest number from a numeric list.

x = int(input("Enter the lenght of the list: "))
lst = []
for i in range(x):
    lst.append(int(input()))

max_num = lst[0]
for i in lst:
    if i > max_num:
        max_num = i
print("Largest number is: ",max_num)
        
```
